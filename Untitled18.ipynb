```python
import pandas as pd
```

```python
df = pd.read_csv('/content/car_prices.csv')
df.head()
```

```python
print("Shape of dataset:", df.shape)
print("Columns in dataset:", df.columns)
```

```python
#Q2 Data Cleaning
# Drop any  row with missing values
df = df.dropna()
#Checking for missing value  are drop or not on   row
df.isnull().sum()
```

```python
# convert data type
df['year'] = pd.to_datetime(df['year'], errors='coerce')
print(df.dtypes)
```

```python
# remove all  duplicate rows
df = df.drop_duplicates()
```

```python
# show the  unique values in the  transmission columns
print(df['transmission'].unique())
```

```python
# Selection B Data analysis with pandas and Numpy
import numpy as np
```

```python
# Q3 Descriptive Analysis
# Add New Columns of prices
random_prices = np.random.randint(200000, 500000, size=len(df))
df['prices'] = random_prices
df.head(10)
```

```python
# find  avrerge,maximum and minimum pirices of cars
average_price =df['prices'].mean()
max_price = df['prices'].max()
min_price = df['prices'].min()


print("Average prices:",average_price)
print("Maximum price:",max_price)
print("minimum price:",min_price)
```

```python
# add new row for Fueltype
Fuel_Types=[' petrol','Diesel','CNG']
df['Fuel_Type'] = np.random.choice(Fuel_Types,size=len(df))
df.head(10)
```

```python
# Most Common fuel type
most_common_fuel_type = df['Fuel_Type']. value_counts().idxmax()
print("Most Common Fuel Type:", most_common_fuel_type)
```

```python
# Count car per transmission type
car_per_transmission = df['transmission'].value_counts()
print("Count of cars per transmission type:")
print(car_per_transmission)
```

```python
# Q4 numpy statistics
# Conver the price column into a numpy array
arr= df['prices'].to_numpy()
print(arr)
```

```python
# Find the mean  median, and standard deviation
mean = np.mean(arr)
median = np.median(arr)
std_dev = np.std(arr)


print("Mean:",mean)
print("Median:",median)
print("Standard Deviation:",std_dev)
```

```python
# Find Number of price above the average
above_average_count = np.sum(arr > mean)
print("Number  of  prices Above the Average:", above_average_count)
```

```python
# plot a bar chart showinng the number of cars by fuel type

# Count each fuel type
fuel_type_counts = df['Fuel_Type'].value_counts()

# Create a bar chart
plt.figure(figsize=(8, 5))
# Use plt.bar directly instead of plt.plt.bar
plt.bar(fuel_type_counts.index, fuel_type_counts.values, color='skyblue', edgecolor='black')


# Add titles and  labels
plt.title('Number of Cars by Fuel Type')
plt.xlabel('Fuel Type')
plt.ylabel('Number of Cars')

# Show the plot
plt.show()
```

```python
# Lineplot chart of average car price per year
# Group by year and calculate average price
average_price_per_year = df.groupby('year')['prices'].mean() # Call the mean() function here

# plot line chart
plt.figure(figsize=(10, 6))
plt.plot(average_price_per_year.index, average_price_per_year.values, marker='o', linestyle='-', color='b')

# Add titles and labels
plt.title('Average Car Price per Year')
plt.xlabel('Year')
plt.ylabel('Average Price')
plt.grid(True, linestyle='--', alpha=0.8)

# Show the plot
plt.show()
```
